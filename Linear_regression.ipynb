# **Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

# **Reading Dataset**

```python
dt = pd.read_csv('Salary_Data.csv')
X = dt.iloc[:, :-1].values
y = dt.iloc[:, -1].values
```

# **Splitting into Training and Test set**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# **Linear regression**

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

# **Predicting Values**

```python
y_pred = regressor.predict(X_test[0,:].reshape(-1,1))
```

```python
print(y_pred)
```

# **Plotting values with Training set**

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train,regressor.predict(X_train),color = 'blue')
plt.title('Salary Expectations vs Years of Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary Expectation')
plt.show()
```

# **Plotting values with Test set**

```python
plt.scatter(X_test,y_test,color = 'red')
plt.plot(X_test,y_pred,color='blue')
plt.title('Salary Expectations vs Years of Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary Expectation')
plt.show()
```
